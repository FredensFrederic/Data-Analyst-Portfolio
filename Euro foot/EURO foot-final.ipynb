```python
import pandas as pd
import matplotlib.pyplot as plt

euro_match = pd.read_csv("/Users/fredo/Documents/SIMPLON DA/Exercice Foot Euro/archive 2/matches/matches/1960.csv")

import glob
import os.path


path_folder_mathes= '/Users/fredo/Documents/SIMPLON DA/Exercice Foot Euro/archive 2/matches/matches'

matches_dfs = []
for file_name in sorted(os.listdir(path_folder_mathes))[:-1]:
    file_path = os.path.join(path_folder_mathes, file_name)
    df_temp = pd.read_csv(file_path)
    print(file_path, df_temp.shape[0])
    matches_dfs.append(df_temp)
df_matches = pd.concat(matches_dfs, axis=0)

print(df_matches.shape)
df_matches

#Delete column not use
only_match = df_matches.drop(columns=['matchday_name','utc_offset_hours','condition_humidity','condition_pitch','condition_temperature','condition_weather','condition_wind_speed','match_attendance','stadium_id','stadium_country_code','stadium_capacity','stadium_latitude','stadium_longitude','stadium_pitch_length','stadium_pitch_width','game_referees', 'stadium_city', 'stadium_name', 'stadium_name_media', 'stadium_name_official', 'stadium_name_event', 'stadium_name_sponsor'])

#Replace team West Germany by Germany
only_match.loc[only_match['away_team'].str.contains('West Germany', case=False), 'away_team'] = 'Germany'
only_match.loc[only_match['home_team'].str.contains('West Germany', case=False), 'home_team'] = 'Germany'

#Delete matches not started yet
only_match = only_match.loc[only_match['status'] == 'FINISHED']
only_match
```

```python
#Number of matches by year
df_matches.groupby('year')['year'].count()
#Number of matches by round
df_matches['round'].value_counts()
```

```python
#Number of goal by Country

home_goal = only_match.groupby(['home_team'])['home_score'].sum()
away_goal = only_match.groupby(['away_team'])['away_score'].sum()
total_goals = pd.concat([home_goal, away_goal], axis=1)
total_goals['home_score'] = total_goals['home_score'].fillna(0)
total_goals['away_score'] = total_goals['away_score'].fillna(0)
total_goals['total'] = total_goals['home_score'] + total_goals['away_score']
total_goals.sort_values('total', ascending=False)
```

```python
#Number of goal conceded by Country

home_conceded = only_match.groupby(['home_team'])['away_score'].sum()
away_conceded = only_match.groupby(['away_team'])['home_score'].sum()
total_conceded = pd.concat([home_conceded, away_conceded], axis=1)
total_conceded['home_score'] = total_conceded['home_score'].fillna(0)
total_conceded['away_score'] = total_conceded['away_score'].fillna(0)
total_conceded['total'] = total_conceded['home_score'] + total_conceded['away_score']
total_conceded.sort_values('total', ascending=False)
```

```python
home_year = only_match.groupby(['year'])['home_score'].sum()
away_year = only_match.groupby(['year'])['away_score'].sum()
total_year = pd.concat([home_year, away_year], axis=1)
total_year['total'] = total_year['home_score'] + total_year['away_score']
total_year.sort_values('total', ascending=False)
year = total_year['total']

plt.bar(year.index, year.values)
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Bar Plot of Most Goals by Year')
plt.show()
```

```python
#Number total of final by Country
number_final_home = only_match.loc[only_match['round'] == 'FINAL'].groupby('home_team')['home_team'].count()
number_final_away =only_match.loc[only_match['round'] == 'FINAL'].groupby('away_team')['away_team'].count()
total_final = pd.concat([number_final_home, number_final_away], axis=1)

total_final['home_team'] = total_final['home_team'].fillna(0)
total_final['away_team'] = total_final['away_team'].fillna(0)
total_final['total'] = total_final['home_team'] + total_final['away_team']
total_final.sort_values('total', ascending=False)
```
